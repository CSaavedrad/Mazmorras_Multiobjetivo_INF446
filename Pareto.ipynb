# Pareto csb

```python
import matplotlib.pyplot as plt
import csv

ruta_csv = "1020_1000_1.5_3000.csv"
```

```python
def obtener_pareto(puntos):
    no_dominados = []
    for i, p in enumerate(puntos):
        dominado = False
        for j, q in enumerate(puntos):
            if j != i:
                if q[0] <= p[0] and q[1] <= p[1] and (q[0] < p[0] or q[1] < p[1]):
                    dominado = True
                    break
        if not dominado:
            no_dominados.append(p)
    return no_dominados
```

```python
puntoscsv = []

# Leer el archivo CSs
with open(ruta_csv, newline='') as archivo:
    lector = csv.reader(archivo, delimiter=';')
    i = 0
    for fila in lector:
        if len(fila) >= 2:  # Asegurarse de que haya al menos dos columnas
            x = float(fila[0])
            y = float(fila[1])
            if x == 1000 and y == 1.5:
                i = i + 1
            puntoscsv.append((x, y))

print(puntoscsv)

puntos = puntoscsv
```

```python
# Lista de puntos (x, y)
#puntos = [
#    (15, 1.75), (15, 1.56), (15, 1.4), (13, 2),
#    (14, 1.86), (16, 1.67), (11, 2), (12, 1.83)
#]



puntos = list(map(lambda p: (abs(p[0] - 1000), abs(p[1] - 1.5)), puntos))

pareto = obtener_pareto(puntos)
print(pareto)
pareto.sort()

x_all, y_all = zip(*puntos)
x_pareto, y_pareto = zip(*pareto)

plt.figure(figsize=(8, 5))
plt.scatter(x_all, y_all, color='xkcd:slate grey', label='Puntos dominados', zorder=2)
plt.scatter(x_pareto, y_pareto, color='xkcd:neon red', label='Frente de Pareto', zorder=3)
plt.plot(x_pareto, y_pareto, color='xkcd:neon red', linestyle='-', label='Conexión Pareto', zorder=3)

title = 'Frente de Pareto - Seed: ' + ruta_csv[0:4]
archive = ruta_csv[0:4] + '.png'

plt.ylim(-0.001, 0.02)     # Ajusta según tu escala de y

plt.title(title)
plt.xlabel('ΔRooms')
plt.ylabel('ΔLinear Coefficient')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(archive, dpi=300, bbox_inches="tight")
plt.show()
```

# Para comparar dos archivos

```python
import matplotlib.pyplot as plt
import csv

ruta_csv1 = "1012_10000_1.5_3000.csv"
ruta_csv2 = "1010_10000_1.5_3000.csv"
```

```python
def obtener_pareto(puntos):
    no_dominados = []
    for i, p in enumerate(puntos):
        dominado = False
        for j, q in enumerate(puntos):
            if j != i:
                if q[0] <= p[0] and q[1] <= p[1] and (q[0] < p[0] or q[1] < p[1]):
                    dominado = True
                    break
        if not dominado:
            no_dominados.append(p)
    return no_dominados
```

```python
puntoscsv1 = []

# Leer el archivo CSs
with open(ruta_csv1, newline='') as archivo:
    lector = csv.reader(archivo, delimiter=';')
    i = 0
    for fila in lector:
        if len(fila) >= 2:  # Asegurarse de que haya al menos dos columnas
            x = float(fila[0])
            y = float(fila[1])
            puntoscsv1.append((x, y))

print(puntoscsv1)

puntos1 = puntoscsv1
```

```python
puntoscsv2 = []

# Leer el archivo CSs
with open(ruta_csv2, newline='') as archivo:
    lector = csv.reader(archivo, delimiter=';')
    i = 0
    for fila in lector:
        if len(fila) >= 2:  # Asegurarse de que haya al menos dos columnas
            x = float(fila[0])
            y = float(fila[1])
            puntoscsv2.append((x, y))

print(puntoscsv2)

puntos2 = puntoscsv2
```

```python
puntos1 = list(map(lambda p: (abs(p[0] - 10000), abs(p[1] - 1.5)), puntos1))
puntos2 = list(map(lambda p: (abs(p[0] - 10000), abs(p[1] - 1.5)), puntos2))

pareto1 = sorted(obtener_pareto(puntos1))
pareto2 = sorted(obtener_pareto(puntos2))

x1, y1 = zip(*puntos1)
xp1, yp1 = zip(*pareto1)

x2, y2 = zip(*puntos2)
xp2, yp2 = zip(*pareto2)

# Gráfico
plt.figure(figsize=(9, 6))

plt.scatter(x1, y1, color='pink', label='Puntos dominados 2-step EA', alpha=0.5)
plt.scatter(x2, y2, color='lightblue', label='Puntos dominados MOEA/D', alpha=0.5)

plt.plot(xp1, yp1, 'r-', label='Frente 2-step EA', zorder=3)
plt.scatter(xp1, yp1, color='red', zorder=3)

plt.plot(xp2, yp2, 'b-', label='Frente Pareto MOEA/D', zorder=3)
plt.scatter(xp2, yp2, color='blue', zorder=3)

plt.ylim(-0.001, 0.06)     # Ajusta según tu escala de y

plt.title('Comparación de Frentes de Pareto')
plt.xlabel('ΔRooms')
plt.ylabel('ΔLinear Coefficient')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("comparacion_pareto.png", dpi=300, bbox_inches="tight")
plt.show()
```
